```python
import rasterio
import numpy as np
from rasterio.merge import merge
from rasterio.plot import show
```

```python
src1 = rasterio.open('image1.tif')
src2 = rasterio.open('image2.tif')
```

```python
show(src1)
```

```python
show(src2)
```

```python
#Putting image 1 and 2 into an array for mosaicking
srcs_to_mosaic = [src1, src2]
```

```python
#Mosaicking images 1 and 2 together.
arr, out_trans = merge(srcs_to_mosaic)
```

```python
show(arr)
```

```python
#Showing the mosaic image using the "terrain" color map
plot.show(arr, cmap='terrain')
```
